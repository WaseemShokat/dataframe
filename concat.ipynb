```python
import pandas as pd
```

```python
us_df=pd.DataFrame({
    'city':['NY','Washington','Hawai'],
    'temperature':[29,22,12],
    'windspeed':[5,7,12]
})
us_df
```

```python
pak_df=pd.DataFrame({
    'city':['Lah','Karac','Fort'],
    'temperature':[30,25,42],
    'windspeed':[6,9,13]
})
pak_df
```

```python
pd.concat([us_df,pak_df])
```

```python
pd.concat([us_df,pak_df],ignore_index=True)
```

```python
df=pd.concat([us_df,pak_df],keys=['Us','Pak'])
df
```

```python
df.loc["Us"]
```

```python
Temp_df=pd.DataFrame({
    'city':['NY','Washington','Hawai'],
    'temperature':[29,22,12],
})
Temp_df
```

```python
wind_df=pd.DataFrame({
    'city':['NY','Washington','Hawai'],
    'windspeed':[5,7,12]
})
wind_df
```

```python
df=pd.concat([Temp_df,wind_df],axis=1)
df
```

```python
wind_df=pd.DataFrame({
    'city':['NY','Washington','Hawai'],
    'windspeed':[5,7,12]
},index=[0,1,2])
wind_df
```

```python
Temp_df=pd.DataFrame({
    'city':['Washington','NY'],
    'temperature':[29,22],
},index=[1,0])
Temp_df
```

```python
df=pd.concat([wind_df,Temp_df],axis=1)
df
```

```python
s=pd.Series(['Rain','Summer','Thunder'],name='event')
s
```

```python
df=pd.concat([wind_df,s],axis=1)
df
```
