```python
def convert_sales(cell):
    if cell==0:
        return 'not given'
    return cell
df1=pd.read_excel("Sponsored Products Purchased product report.xlsx", converters={
    "7 Day Other SKU Units (#)":convert_sales,
    "7 Day Other SKU Orders (#)":convert_sales,
    "7 Day Other SKU Sales":convert_sales
})
df1
```
